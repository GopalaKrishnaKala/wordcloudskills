```python
# Generate word cloud with specified colors: Indian Red, Dark Blue, Orange, Green
# Define custom color function
from wordcloud import WordCloud
import matplotlib.pyplot as plt
def custom_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    colors = ['#CD5C5C', '#00008B', '#FFA500', '#008000']  # Indian Red, Dark Blue, Orange, Green
    return colors[hash(word) % len(colors)]
skills = [
    "Python", "R", "Java", "C", "C++", "Javascript", "Shell Scripting", "Linux",
    "UNIX", "AWS EC2", "S3", "Lambda", "Azure", "Data Management", 
    "Visualization", "ETL", "SQL", "NoSQL", "MongoDB", "Hive", "Kafka", "Spark", 
    "Databricks", "Tableau", "PowerBI", "Data Science", "Machine Learning", 
    "Pandas", "Numpy", "Matplotlib", "Seaborn", "Scikit-Learn", "TensorFlow", 
    "PyTorch", "PySpark", "NLP"
]
# Generate word cloud with custom colors
wordcloud_custom_colors = WordCloud(width=1200, height=600, background_color='white', color_func=custom_color_func).generate(' '.join(skills))

# Display the word cloud with custom colors
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud_custom_colors, interpolation='bilinear')
plt.axis('off')
plt.show()
```
